```python
#Drifted Brownian motion simulator, set sigma, mu, number of timesteps
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import imageio
import gif_2d_Brown as gif
import importlib

importlib.reload(gif)

steps=1001 #timesteps
mu=0.01 #drift
time=np.linspace(0,1,steps)
sigma=0.1
W=np.linspace(0,0,steps)
W[0]=100
#print(W)
#print(time)
for i in range (1000):
    #print(W[i],W[i+1])
    W[i+1]=W[i]+W[i]*mu*1/1000+W[i]*sigma*np.random.normal(0,np.sqrt(1/(steps-1)))
    #print(W[i],W[i+1])
#print(W)
```

```python
plt.plot(time, W)  # Use markers to highlight data points
plt.xlabel('Time')
plt.ylabel('W')
plt.title('Time vs W')
plt.grid(True)
plt.show()
```
