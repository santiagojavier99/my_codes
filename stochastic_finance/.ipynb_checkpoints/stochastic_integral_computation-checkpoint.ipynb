```python
#using linear increment of the partion
import numpy as np
import matplotlib.pyplot as plt

def brownian_matrix(M,N,T):                 #function that defines the M brownian trajectories 
    W=np.zeros((M,N), dtype='float')
    for i in range (M):
        for j in range (N-1):
            W[i,j+1]=W[i,j]+np.random.normal(0, np.sqrt(T/(N-1)))
    return W

finest_part=100
simulation=np.linspace(0,finest_part,finest_part)
T=1                        #setting time T
M=20                       #setting the number of Brownian trajectories
error=[]
error.append(0)            #initializing the firt two elements since they correspond to no  partitions
error.append(0)
for N in range (2,finest_part):   #for over the partitions 
    time=np.linspace(0,T,N)
    W=brownian_matrix(M,N,T)    
    #print(W)
    estimator=[]                
    for i in range (M):
        sum=0
        for j in range(N-1):
            sum=sum+ W[i,j]*(W[i,j+1]-W[i,j])
        estimator.append(((W[i,N-1]**2-T)/2-sum)**2)
    error.append(np.average(estimator))       #computation of the error for each partition 

#plotting the error VS the partion
plt.plot(simulation, error)  
plt.xlim(2,100)
plt.xlabel('partition')
plt.ylabel('error')
plt.title('error vs partition')
plt.grid(True)
plt.show()

        
    
```

```python
#considering the partition with exponential increment 
import numpy as np
import matplotlib.pyplot as plt

def brownian_matrix(M,N,T):                # function that defines the M Brownian trajectories
    W = np.zeros((M,N), dtype='float')
    for i in range(M):
        for j in range(N-1):
            W[i,j+1]=W[i,j]+np.random.normal(0,np.sqrt(T/(N-1)))
    return W

n=10                                        # number of considered partitions
simulation=[2**(i+1)for i in range(n)]      # exponential increment of partitions
T=1                                         # setting time T

# Different values of M for comparison
M_values=[2,10,100]

plt.figure(figsize=(10,6))

for M in M_values:
    error=[]
    for N in simulation:                      # for each partition size
        W = brownian_matrix(M, N, T)
        estimator = []
        for i in range(M):
            sum = 0
            for j in range(N - 1):
                sum=sum+W[i,j]*(W[i,j+1]-W[i,j])
            estimator.append(((W[i,N-1]**2-T)/2-sum)**2)
        
        error.append(np.average(estimator))                   # average error for each partition size

    # Plot each M on the same graph with a label
    plt.plot(simulation, error, marker='o', label=f'M = {M}')   #adding to the plot 

# Final plot settings
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Partition (log scale)')
plt.ylabel('Error (log scale)')
plt.title('Error vs Partition (Log Scale) for Different M Values')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
#NOTE: increasing the number M of Brownian trajectories considered improves significatly the error behaviour: log(error) \sim log(N)
```
